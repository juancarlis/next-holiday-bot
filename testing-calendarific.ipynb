```python
import pandas as pd
import requests
import datetime
import json
```

```python
country = 'AR'
year = '2021'
API_KEY = ''

response = requests.get(f'https://calendarific.com/api/v2/holidays?&api_key={API_KEY}&country={country}&year={year}')
data = json.loads(response.text)
```

```python
data
```

```python
clean_data = data['response']['holidays']
```

```python
df = pd.DataFrame.from_dict(clean_data)
```

```python
df
```

```python
df.loc[0, 'date']['datetime']
```

```python
df['country'] = df['country'].apply(lambda x: x['name'])
```

```python
df['date'] = df['date'].apply(lambda x: pd.Timestamp(
                year=x['datetime']['year'],
                month=x['datetime']['month'],
                day=x['datetime']['day']))
```

```python
df['type'] = df['type'].apply(lambda x: ','.join(x))
```
